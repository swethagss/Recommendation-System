## *Basic Recommendation System*

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies_df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
print(f"Credits has {credits.shape[0]} no of rows and {credits.shape[1]} no of columns")
print(f"Movies_df has {movies_df.shape[0]} no of rows and {movies_df.shape[1]} no of columns")
```

### Combining both datasets

```python
credits_column_renamed = credits.rename(index = str, columns = {'movie_id' : 'id'})
movies_df_merge = movies_df.merge(credits_column_renamed, on = 'id')
movies_df_merge.head()
```

```python
## Dropping unnecessary columns

movies_cleaned_df = movies_df_merge.drop(columns=['homepage', 'production_countries','status',
                                                  'title_x', 'title_y'])

movies_cleaned_df.head()
```

```python
movies_cleaned_df.info()
```

### Calculating all the components based on the formula



```python
v = movies_cleaned_df['vote_count']
R = movies_cleaned_df['vote_average']
C = movies_cleaned_df['vote_average'].mean()
m = movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
movies_cleaned_df['weighted_average'] = ((R * v) + (C * m))/ (v+m)
```

```python
movies_cleaned_df.head()
```

```python
movies_sorted_ranking = movies_cleaned_df.sort_values('weighted_average', ascending = False)
movies_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(15)
```

```python
weight_average=movies_sorted_ranking.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(10), y=weight_average['original_title'].head(10), data=weight_average)
#plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
popularity = movies_sorted_ranking.sort_values('popularity', ascending = False)
plt.figure(figsize = (10,6))
ax = sns.barplot(x = popularity['popularity'].head(10), y = popularity['original_title'].head(10),
                 data = popularity)

plt.title("Popular movies based on Votes", weight = 'bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_popular_movies.png')
```

```python
popularity.head()
```

#### Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movies_cleaned_df[['weighted_average', 'popularity']])
movies_normalized_df = pd.DataFrame(movie_scaled_df, columns = ['weighted_average', 'popularity'])

movies_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weighted_avg', 'normalized_popularity']] = movies_normalized_df

```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score'] = (movies_cleaned_df['normalized_weighted_avg'] * 0.5 + 
                              movies_cleaned_df['normalized_popularity'] * 0.5)


movies_scored_df = movies_cleaned_df.sort_values(['score'], ascending = False)
movies_scored_df[['original_title', 'normalized_weighted_avg', 
                  'normalized_popularity', 'score']].head()

```

```python
plt.figure(figsize=(10,6))

ax = sns.barplot(x = movies_scored_df['score'].head(10), y = movies_scored_df['original_title'].head(10))

plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

plt.savefig('scored_movies.png')
                 
```


## Content Based Recommendation System

##### Now lets make a recommendations based on the movie’s plot summaries given in the overview column. So if our user gives us a movie title, our goal is to recommend movies that share similar plot summaries.

```python
movies_cleaned_df.head(1)['overview']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

# Filling NaNs with empty string
movies_cleaned_df['overview'] = movies_cleaned_df['overview'].fillna('')
```

```python
# Fitting the TF-IDF on the 'overview' text
tfv_matrix = tfv.fit_transform(movies_cleaned_df['overview'])
```

```python

tfv_matrix.shape
```

```python

from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movies_cleaned_df.index, index=movies_cleaned_df['original_title']).drop_duplicates()
```

```python
indices
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return movies_cleaned_df['original_title'].iloc[movie_indices]
```

```python
# Testing our content-based recommendation system with the seminal film Spy Kids
give_rec('Spy Kids')
```
