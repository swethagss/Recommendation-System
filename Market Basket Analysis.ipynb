## Implemening Market Basket Analysis

```python
import pandas as pd
import numpy as np

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_excel('Online Retail.xlsx')
```

### *Data Preparation*

```python
data['Description'] = data['Description'].str.strip()  ##removing white spaces
data.dropna(axis = 0, subset=['InvoiceNo'], inplace = True) ##removing duplicates
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
data = data[~data['InvoiceNo'].str.contains('C')]
data.head()
```

```python
data['Country'].value_counts()
```

```python
## Doing Analysis for one country so seperating transactions for Germany

mybasket = (data[data['Country'] == 'Germany'].
            groupby(['InvoiceNo', 'Description'])['Quantity']
            .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
mybasket.head()
```

```python
## converting all positive values to 1 and rest as 0

def my_encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1


my_basket_sets = mybasket.applymap(my_encode_units)
my_basket_sets.drop('POSTAGE', inplace = True, axis =1)  ## removing POSTAGE column
```

## Model Training

```python
## Generating frequent item sets

my_frequent_itemsets = apriori(my_basket_sets, min_support = 0.07, use_colnames=True)
```

```python
## generating rules
my_rules = association_rules(my_frequent_itemsets, metric = 'lift', min_threshold=1)
```

```python
## viewing top 100 rules
my_rules.head(100)
```
